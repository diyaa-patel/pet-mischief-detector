```python
# data download
import fiftyone as fo
import fiftyone.zoo as foz

# Choose your classes
my_classes = ["cat", "dog", "cup", "laptop", "potted plant", "book", "tv", "remote", "frisbee", "vase"]

# This downloads the metadata and the specific images
dataset = foz.load_zoo_dataset(
    "coco-2017",
    splits=["train", "validation"],
    classes=my_classes,
    max_samples=600,
)
```

```python
# Task 2: split dataset into train / val / test and export in YOLO format

import os
import random
import yaml
from collections import Counter

SEED = 42
EXPORT_DIR = "./pet_mischief_data"

TRAIN_RATIO = 0.8
VAL_RATIO = 0.1
TEST_RATIO = 0.1

# Remove old tags if you rerun this cell
dataset.untag_samples(["train", "val", "test"])

# Get all sample IDs and shuffle with fixed seed
sample_ids = dataset.values("id")
random.Random(SEED).shuffle(sample_ids)

# Compute split sizes
n_total = len(sample_ids)
n_train = int(n_total * TRAIN_RATIO)
n_val = int(n_total * VAL_RATIO)
n_test = n_total - n_train - n_val

train_ids = sample_ids[:n_train]
val_ids = sample_ids[n_train:n_train + n_val]
test_ids = sample_ids[n_train + n_val:]

# Tag samples
dataset.select(train_ids).tag_samples("train")
dataset.select(val_ids).tag_samples("val")
dataset.select(test_ids).tag_samples("test")

print(f"Total samples: {n_total}")
print(f"Train: {len(train_ids)}")
print(f"Val:   {len(val_ids)}")
print(f"Test:  {len(test_ids)}")

# count labels in each split
def count_labels(view, label_field="ground_truth"):
    counts = Counter()
    for sample in view:
        if sample[label_field] is not None:
            for det in sample[label_field].detections:
                if det.label in my_classes:
                    counts[det.label] += 1
    return counts

train_view = dataset.match_tags("train")
val_view = dataset.match_tags("val")
test_view = dataset.match_tags("test")

print("\nTrain label counts:", count_labels(train_view))
print("Val label counts:  ", count_labels(val_view))
print("Test label counts: ", count_labels(test_view))

# Export each split separately in YOLO format
for split_name, view in [("train", train_view), ("val", val_view), ("test", test_view)]:
    view.export(
        export_dir=EXPORT_DIR,
        dataset_type=fo.types.YOLOv5Dataset,
        label_field="ground_truth",
        split=split_name,
        classes=my_classes
    )

# Create dataset.yaml for later YOLO training
yaml_data = {
    "path": os.path.abspath(EXPORT_DIR),
    "train": "images/train",
    "val": "images/val",
    "test": "images/test",
    "names": {i: cls for i, cls in enumerate(my_classes)}
}

with open(os.path.join(EXPORT_DIR, "dataset.yaml"), "w") as f:
    yaml.dump(yaml_data, f, sort_keys=False)

print("\nExport complete.")
print("Saved to:", os.path.abspath(EXPORT_DIR))
```

```python
# Task 3: Fine-tune YOLOv8 on the Pet Mischief Dataset
from ultralytics import YOLO
import os

# --- CONFIGURATION (Edit these as needed) ---
# 1. Path to your dataset.yaml file created in Task 2
# Ensure the 'pet_mischief_data' folder is uploaded to your Colab files!
DATASET_CONFIG = "/content/pet_mischief_data/dataset.yaml"

# 2. Select the pretrained model weights (Transfer Learning)
# 'yolov8n.pt' is the Nano version (fastest).
# You could use 'yolov8s.pt' (Small) for slightly better accuracy if you have time.
MODEL_NAME = 'yolov8n.pt'

# 3. Hyperparameters
EPOCHS = 50          # Number of passes through the data
IMG_SIZE = 640       # Standard resolution for YOLOv8
BATCH_SIZE = 16      # Number of images processed at once (adjust based on GPU memory)
PROJECT_FOLDER = "/content/drive/MyDrive/Pet_Mischief_Project/results"
RUN_NAME = 'pet_mischief_detector_final'

# --- TRAINING EXECUTION ---

def main():
    # Load the pretrained model
    print(f"Loading pretrained model: {MODEL_NAME}...")
    model = YOLO(MODEL_NAME)

    # Start training
    print(f"Starting training on {DATASET_CONFIG}...")
    results = model.train(
        data=DATASET_CONFIG,
        epochs=EPOCHS,
        imgsz=IMG_SIZE,
        batch=BATCH_SIZE,
        project=PROJECT_FOLDER,
        name=RUN_NAME,
        device=0,             # Uses GPU (set to 'cpu' if no GPU is available)
        save=True,            # Save checkpoints and best model
        plots=True            # Generate the Confusion Matrix and Loss curves for your report
    )

    print("\n" + "="*30)
    print("TRAINING COMPLETE")
    print(f"Weights and logs saved to: {PROJECT_FOLDER}/{RUN_NAME}")
    print("="*30)

if __name__ == "__main__":
    main()
```

```python
# Task 4B: Mischief logic + qualitative evaluation

import os
import math
import cv2
from ultralytics import YOLO

RESULTS_DIR = "/content/drive/MyDrive/Pet_Mischief_Project/results"
TEST_IMAGES_DIR = "/content/pet_mischief_data/images/test"
OUTPUT_DIR = "/content/mischief_outputs"

os.makedirs(OUTPUT_DIR, exist_ok=True)

best_models = []
for root, dirs, files in os.walk(RESULTS_DIR):
    if "best.pt" in files:
        best_models.append(os.path.join(root, "best.pt"))

if not best_models:
    raise FileNotFoundError("No best.pt file found in the results folder.")

BEST_MODEL_PATH = max(best_models, key=os.path.getmtime)
print("Using model:", BEST_MODEL_PATH)

model = YOLO(BEST_MODEL_PATH)

# Risk weights for objects
OBJECT_RISK = {
    "cup": 35,
    "laptop": 50,
    "potted plant": 45,
    "book": 15,
    "tv": 40,
    "remote": 30,
    "frisbee": 20,
    "vase": 55
}

# Confidence thresholds
PET_CONF_THRESHOLD = 0.50
OBJECT_CONF_THRESHOLD = 0.40

def box_iou(box1, box2):
    x1 = max(box1[0], box2[0])
    y1 = max(box1[1], box2[1])
    x2 = min(box1[2], box2[2])
    y2 = min(box1[3], box2[3])

    inter_w = max(0, x2 - x1)
    inter_h = max(0, y2 - y1)
    inter_area = inter_w * inter_h

    area1 = max(0, box1[2] - box1[0]) * max(0, box1[3] - box1[1])
    area2 = max(0, box2[2] - box2[0]) * max(0, box2[3] - box2[1])

    union = area1 + area2 - inter_area
    if union == 0:
        return 0.0

    return inter_area / union

def center_distance(box1, box2):
    c1x = (box1[0] + box1[2]) / 2
    c1y = (box1[1] + box1[3]) / 2
    c2x = (box2[0] + box2[2]) / 2
    c2y = (box2[1] + box2[3]) / 2
    return math.sqrt((c1x - c2x) ** 2 + (c1y - c2y) ** 2)

def calculate_mischief(detections, image_w, image_h):
    pets = [d for d in detections if d["label"] in ["cat", "dog"]]
    objects = [d for d in detections if d["label"] not in ["cat", "dog"]]

    if len(pets) == 0:
        return 0, "LOW RISK: No pets detected."

    if len(objects) == 0:
        return 10, "LOW RISK: Pet detected, but no risky object nearby."

    img_diag = math.sqrt(image_w ** 2 + image_h ** 2)

    best_score = 0
    best_message = "LOW RISK: Pet detected."

    for pet in pets:
        for obj in objects:
            obj_label = obj["label"]
            obj_weight = OBJECT_RISK.get(obj_label, 10)

            dist = center_distance(pet["box"], obj["box"])
            normalized_dist = dist / img_diag
            proximity_score = max(0, 1 - normalized_dist)

            iou = box_iou(pet["box"], obj["box"])
            overlap_bonus = 30 if iou > 0 else 0

            pair_score = obj_weight * (0.5 + proximity_score) + overlap_bonus

            if pair_score > best_score:
                best_score = pair_score
                pet_name = pet["label"]

                if pair_score >= 65:
                    best_message = f"HIGH RISK: {pet_name} is very close to the {obj_label}!"
                elif pair_score >= 40:
                    best_message = f"MEDIUM RISK: {pet_name} may interact with the {obj_label}."
                else:
                    best_message = f"LOW RISK: {pet_name} is not too close to the {obj_label}."

    best_score = min(100, round(best_score))
    return best_score, best_message

image_files = [
    f for f in os.listdir(TEST_IMAGES_DIR)
    if f.lower().endswith((".jpg", ".jpeg", ".png"))
]

summary = []

for image_name in image_files:
    image_path = os.path.join(TEST_IMAGES_DIR, image_name)

    results = model(image_path, conf=0.25, verbose=False)
    r = results[0]

    img = cv2.imread(image_path)
    h, w = img.shape[:2]

    detections = []
    kept_boxes = []

    for box in r.boxes:
        cls_id = int(box.cls[0].item())
        conf = float(box.conf[0].item())
        label = model.names[cls_id]
        x1, y1, x2, y2 = box.xyxy[0].tolist()

        # Filter out weak detections
        if label in ["cat", "dog"] and conf < PET_CONF_THRESHOLD:
            continue
        if label not in ["cat", "dog"] and conf < OBJECT_CONF_THRESHOLD:
            continue

        detections.append({
            "label": label,
            "conf": conf,
            "box": [x1, y1, x2, y2]
        })

        kept_boxes.append(box)

    risk_score, warning = calculate_mischief(detections, w, h)

    # Start from original image
    plotted = img.copy()

    # Draw only filtered boxes
    for det in detections:
        x1, y1, x2, y2 = map(int, det["box"])
        label_text = f'{det["label"]} {det["conf"]:.2f}'

        if det["label"] in ["cat", "dog"]:
            color = (255, 0, 0)   # blue-ish in OpenCV BGR
        else:
            color = (255, 255, 0) # cyan-ish

        cv2.rectangle(plotted, (x1, y1), (x2, y2), color, 2)
        cv2.putText(
            plotted,
            label_text,
            (x1, max(y1 - 10, 20)),
            cv2.FONT_HERSHEY_SIMPLEX,
            0.7,
            color,
            2
        )

    # Add risk score and warning text
    cv2.putText(
        plotted,
        f"Risk Score: {risk_score}/100",
        (20, 35),
        cv2.FONT_HERSHEY_SIMPLEX,
        1,
        (0, 0, 255),
        2
    )

    cv2.putText(
        plotted,
        warning,
        (20, 75),
        cv2.FONT_HERSHEY_SIMPLEX,
        0.8,
        (255, 255, 255),
        2
    )

    output_path = os.path.join(OUTPUT_DIR, image_name)
    cv2.imwrite(output_path, plotted)

    summary.append((image_name, risk_score, warning))

print("Saved annotated outputs to:", OUTPUT_DIR)
print("\nSample results:")
for row in summary[:10]:
    print(row)
```
